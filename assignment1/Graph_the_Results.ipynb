```python
# ACCURACY BY NUMBER OF ITERATIONS for each ALGORITHM =============


# Create a loop andsave the resulting graphs
models = ['KNN']#['ANN', 'Boost', 'DT', 'KNN', 'SVM_Lin', 'SVM_RBF']
datasets = ['wine'] #['credit'] #, 'wine']
#portions = ['training', 'test':

for model in models:
    for s in datasets:
        #for p in portions:
        lc_test = []
        lc_train = []

        # get data
        file_path3 ="./output/"
        file_train = model+'_'+s+'_LC_train.csv' 
        file_test = model+'_'+s+'_LC_test.csv'

        lc_test = pd.read_csv (file_path3+file_train, sep =",")
        lc_train = pd.read_csv (file_path3+file_test, sep =",")
        
        if s=='credit':
            # Get avg of each row (mean of TRAIN accuaracy over all folds for credit)
            dfs = list()
            for l in range(0, len(lc_train['0'])):
                dfs.append(lc_train[['0', '1', '2', '3', '4']].iloc[l].mean(axis=0))
            # change list to np array
            mean = np.asarray(dfs, dtype=np.float32)
            # add array to df
            lc_train['mean'] = mean


            # Get avg of each row (mean of TEST accuaracy over all folds for credit dataset)
            dfs = list()
            for l in range(0, len(lc_test['0'])):
                dfs.append(lc_test[['0', '1', '2', '3', '4']].iloc[l].mean(axis=0))
            # change list to np array
            mean = np.asarray(dfs, dtype=np.float32)
            # add array to df
            lc_test['mean'] = mean
            
        if s== 'wine':
               # Get avg of each row (mean of TRAIN accuaracy over all folds for wine dataset)
                dfs = list()
                for l in range(0, len(lc_train['0'])):
                    dfs.append(lc_train[['0', '1', '2']].iloc[l].mean(axis=0))
                # change list to np array
                mean = np.asarray(dfs, dtype=np.float32)
                # add array to df
                lc_train['mean'] = mean


                # Get avg of each row (mean of TEST accuaracy over all folds for wine dataset)
                dfs = list()
                for l in range(0, len(lc_test['0'])):
                    dfs.append(lc_test[['0', '1', '2']].iloc[l].mean(axis=0))
                # change list to np array
                mean = np.asarray(dfs, dtype=np.float32)
                # add array to df
                lc_test['mean'] = mean 
            
            
        # train accuracy and test accuracy by iteration together in same df to plot
        acc = pd.DataFrame(columns = ('Iterations', 'Test Accuracy', 'Train Accuracy'))
        acc['Iterations'] = lc_train['Unnamed: 0']
        acc['Test Accuracy'] = lc_test['mean']
        acc['Train Accuracy'] = lc_train['mean']

        # plot the results

        # style
        plt.style.use('seaborn-darkgrid')

        # create a color palette
        palette = plt.get_cmap('Set1')

        # multiple line plot
        num=0
        for column in acc.drop('Iterations', axis=1):
            num+=1
            plt.plot(acc['Iterations'], acc[column], marker='o', 
                     color=palette(num), linewidth=2, alpha=0.9, label=column)

        #     numz = 0
        #     for column in rbf_curve.drop('param_SVM__max_iter', axis=1):
        #         numz+=1
        #         plt.plot(rbf_curve['param_SVM__max_iter'], rbf_curve[column], marker='', 
        #                  color=palette(num), linewidth=2, alpha=0.9, label='rbf '+column)

        # Add legend
        plt.legend(loc='best', ncol=1)

        # Add titles
        plt.title(model+" Accuracy by Iterations on "+s+' data', loc='center', fontsize=12, fontweight=0, color='darkblue')
        plt.xlabel("Iterations")
        plt.ylabel("Accuracy")
        #plt.xlim(xmin=0, xmax=6000)
        plt.savefig('images/'+model+'_'+s+'_AccByIterationsII.png')
        plt.show()
```

```python
# PRUNING ANALYSIS: ======================================
# Number of Internal Nodes at different values of alphas
# C = 1-alpha and C is the weights of the score of the tree, with hightest 
#   scores being kept & lower scores pruned from the tree

# get data
dataset = ['wine', 'credit']

for ds in dataset: 
    file_path3 ="./output/"
    file = 'DT_'+ds+'_nodecounts.csv' 

    nodes = pd.read_csv (file_path3+file, sep =",", header=None)
    # changing columns using .columns() 
    nodes.columns = ['alpha', 'nodes'] 

    # plot all the values
    nodes.plot(x='alpha', y='nodes', marker='o')
    plt.title('Nodes for Different Alphas - '+ds+' data', loc='center', fontsize=12, fontweight=0, color='darkblue')
    plt.xlabel("Pruning Alpha Values")
    plt.ylabel("# of Internal Nodes")
    plt.legend()
    plt.savefig('images/DT_'+ds+'_numnodes.png')
    plt.show()

    # zoom in on the curve
    nodes2 = nodes.iloc[4:12]
    nodes2.plot(x='alpha', y='nodes', marker='o')
    plt.title('Closer Look: Nodes for Different Alphas - '+ds+' data', loc='center', fontsize=12, fontweight=0, color='darkblue')
    plt.xlabel("Pruning Alpha Values")
    plt.ylabel("# of Internal Nodes")
    plt.legend()
    plt.savefig('images/DT_'+ds+'_numnodes_zoom.png')
    plt.show()

    plt.show()
```

```python
# ANN - Plot Average accuracy by hidden layer dimensions =================================

# style of viz
plt.style.use('seaborn-darkgrid')
# create a color palette
palette = plt.get_cmap('Set1')

models = ['ANN']
dataset = ['wine', 'credit']
for model in models:
    for ds in dataset:
        # get the data
        #file_path3 ="//Users//jamieandrews//Documents//CS_7641//Assignment1//JonTay//output//"
        #file_path3 ="//Users//jamieandrews//Documents//CS_7641//Assignment1//JonTay//output//"
        
        file = './output/'+model+'_'+ds+'_reg.csv' 

        reg = pd.read_csv (file, sep =",")
        # change hidden layers to categorical variable
        reg['param_MLP__hidden_layer_sizes'] = reg['param_MLP__hidden_layer_sizes'].astype('category')
        # get mean test and train scores grouping by hidden layer
        reg_byHL = reg.groupby(['param_MLP__hidden_layer_sizes'])['mean_train_score','mean_test_score'].mean()
        # sort values before plotting
        reg_byHL = reg_byHL.sort_values('mean_test_score')
        # plot it
        reg_byHL.plot.bar()
        plt.title("ANN Accuracy by Hidden Layer Sizes - "+ds+" data", loc='center', fontsize=12, fontweight=0, color='darkblue')
        plt.xlabel("Hidden Layer Dimensions")
        plt.ylabel('Accuracy Score')
        plt.legend(loc='best')
        plt.xticks(rotation=60)
        plt.ylim(ymax = 0.7, ymin =0)
        plt.savefig('./images/ANN_'+ds+'_acc_layers.png')
        plt.show()
        #plt.close()
```

```python
# GRAPH DT NODES AND ACCURACY BY ALPHAS === dual y-axis graph =======

import numpy as np
import matplotlib.pyplot as plt

# GRAPH DT NODES AND ACCURACY BY ALPHAS
datasets = ['wine', 'credit']
for ds in datasets:
    # get the data
    file = './output/DT_'+ds+'_prune_acc.csv'
    alphas = pd.read_csv (file, sep =",", header=0)
    # rename the columns
    alphas.columns = ['ID', 'Dataset', 'Alpha' , 'Nodes', 'Accuracy'] 
    alphas.drop(['ID'], axis=1, inplace=True)

    # style of viz
    plt.style.use('seaborn-darkgrid')
    # create a color palette
    palette = plt.get_cmap('Set1')


    fig, ax1 = plt.subplots()
    # t is the x axis; nodes go here
    t = alphas['Alpha']
    s1 = alphas['Nodes']
    ax1.plot(t, s1, 'bo')
    ax1.set_xlabel('Alphas')

    # Make the y-axis label, ticks and tick labels match the line color.
    ax1.set_ylabel('Number of Internal Nodes', color='b')
    ax1.tick_params('y', colors='b')

    ax2 = ax1.twinx()
    s2 = alphas['Accuracy']
    ax2.plot(t, s2, 'ro')
    #plt.xlim(xmax =.02, xmin =-.02)
    ax2.set_ylabel('Accuracy', color='red')
    ax2.tick_params('y', colors='red')
    #plt.legend(loc='best')
    plt.title(model+" Accuracy & Nodes by Alpha Values, "+ds+" data", loc='center', fontsize=12, fontweight=0, color='darkblue')
    plt.savefig('./images/DT_'+ds+'_AccNodesAlphas.png')      
    fig.tight_layout()
    plt.show()
```
